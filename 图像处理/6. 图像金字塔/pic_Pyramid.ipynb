```python
import cv2
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
img = cv2.imread("./pic/jinx.jpeg")

def cv_show(img,name):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
cv_show(img,"jinx")
```

```python
img.shape
```

```python
up = cv2.pyrUp(img)
# cv_show(up,"up")
plt.imshow(up)
up.shape
```

```python
down = cv2.pyrDown(img)
plt.imshow(down)
down.shape
```

```python
twice = cv2.pyrDown(up)
plt.imshow(twice)
```

```python
res = np.hstack((twice,img))
cv_show(res,"res")
plt.imshow(res)
#里外里损失了两次，清晰度降低了
```

### 拉普拉斯金字塔
![](pic/拉普拉斯金字塔.png)

```python
laplace = img - cv2.pyrUp(cv2.pyrDown(img))
cv_show(laplace,"laplace")
plt.imshow(laplace)
```
