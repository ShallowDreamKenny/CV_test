```python
import cv2
import matplotlib.pylab as plt
import numpy as np

img = cv2.imread("./pic/car.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
V1 = cv2.Canny(img,80,150)
V2 = cv2.Canny(img,50,100)
res = np.hstack((V1,V2))
```

```python
def cv_show(img,name):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
cv_show(res,"res")
```

```python
V1 = cv2.cvtColor(V1,cv2.COLOR_BAYER_GR2RGB)
V2 = cv2.cvtColor(V2,cv2.COLOR_BAYER_GR2RGB)
plt.subplot(121),plt.imshow(V1)
plt.subplot(122),plt.imshow(V2)
```

## 阈值大一些检测的严格，即边框少一些
## 阈值小一些，检测的更送，即边框多一些
